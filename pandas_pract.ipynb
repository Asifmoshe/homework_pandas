```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Evan"],
    "Age": [25, 30, 35, 28, 40],
    "Salary": [50000, 60000, 75000, 62000, 80000],
    "Department": ["HR", "IT", "Finance", "IT", "HR"]
}

df = pd.DataFrame(data)
print(df)
```

```python
result = df.iloc[0:3, [0, 2]]
display(result)
```

```python
df['Bonus'] = 0.1 * df['Salary']
display(df)
```

```python
df = df.set_index('Name')
display(df)
```

```python
result = df.loc['Bob':'Diana', ['Salary', 'Department']]
display(result)
```

```python
df = df.reset_index()
display(df)
```
